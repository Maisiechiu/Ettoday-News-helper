```python
import pandas as pd

# 读取TSV文件
df = pd.read_csv('/mnt/sda/maisie/Ettoday-News-helper/ettoday_news.tsv', sep='\t')

# 选择所需的字段
selected_data = df[['cat_lv1_id', 'cat_lv1', 'cat_lv2_id', 'cat_lv2']]

result_dict = {}

for index, row in selected_data.iterrows():
    key = f"{row['cat_lv1_id']}-{row['cat_lv2_id']}"
    value = f"{row['cat_lv1']}-{row['cat_lv2']}"
    result_dict[key] = value

# 打印结果
for key, value in result_dict.items():
    print(f"{key}: {value}")
```
